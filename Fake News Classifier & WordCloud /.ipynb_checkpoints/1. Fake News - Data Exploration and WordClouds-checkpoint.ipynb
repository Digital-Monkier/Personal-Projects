### Digging into the statistical features of the news titles

```python
import numpy as np
import string

fake_filt = news[news['label'] == 'FAKE'] 
real_filt = news[news['label'] == "REAL"]

fake_length = [len(x) for x in fake_filt['title']]
real_length = [len(x) for x in real_filt['title']]

print('Average Fake Length:',sum(fake_length) / len(fake_length))
print('Max Fake Length:', max(fake_length))
print('Min Fake Length:', min(fake_length), '\n')
print('Average Real Length:',sum(real_length) / len(real_length))
print('Max Real Length:', max(real_length))
print('Min Real Length:', min(real_length))
```

```python
def punct_percent(txt):
    count = sum([1 for x in txt if x in string.punctuation])
    return 100*count/len(txt)
```

```python
fake_percent = fake_filt['title'].apply(lambda x: punct_percent(x))
real_percent = real_filt['title'].apply(lambda x: punct_percent(x))
print('Average Punctuation In Fake News Titles:', sum(fake_percent)/len(fake_filt['title']))
print('Average Punctuation In Real News Titles:', sum(real_percent)/len(real_filt['title']))
```

```python
real_trumps = real_filt[real_filt['title'].str.contains("Trump")]
fake_trumps = fake_filt[fake_filt['title'].str.contains("Trump")]

print(real_trumps['title'].head())
print(fake_trumps['title'].head())
```

```python
# Generate the word cloud from the east_of_eden string
all_cloud = WordCloud(background_color="white",  font_path = '/System/Library/Fonts/HelveticaNeue.ttc', 
                       stopwords=my_stopwords, width=800, height=400).generate(' '.join(news['text']))
plt.figure(figsize=(20,10))
# Create a figure of the generated cloud
plt.imshow(all_cloud, interpolation='bilinear')  
plt.axis('off')
# Display the figure
plt.show()
```

### Creating a mini df using a wordlist and their counts in the filtered REAL and FAKE article
The idea here is to use the counts as a condition for the custom color func within wordcloud as a visual representation as to which of the words are more common in REAL or FAKE.

```python
wordlist = ['Russia', 'Russian','Donald', 'Sex', 'Scandal', 'Republican', 'Leftist', 'Administration', 'Conservative',
            'Trump', 'Hillary', 'Clinton', 'DNC', 'GOP', 'Congress', 'Twitter', 'Fox', 'News', 'Treason',
            'Putin', 'Facebook', 'Hack', 'Hacker', 'Hacking', 'Election', 'Ivanka', 'Bill', 'Law', 'leftist', 'United',
            'States', 'White', 'House', 'media', 'policy', 'affairs', 'Celebrity', 'video', 'shocking', 'American', 
            'Government', 'Mitch']


fake_text = fake_filt['title']
fake_join = " ".join(fake_text)
fake_split = fake_join.split(" ")

real_text = real_filt['title']
real_join = " ".join(real_text)
real_split = real_join.split(" ")

fake_freq = [fake_split.count(w) for w in wordlist]
real_freq = [real_split.count(w) for w in wordlist]

mini_df = pd.DataFrame(list(zip(wordlist, fake_freq, real_freq)), columns = ['name', 'fake_freq', 'real_freq'])

mini_df.sort_values(by='fake_freq',ascending=False)
```

```python
mini_df.loc[mini_df['fake_freq'] > mini_df['real_freq'], 'fakes'] = mini_df['name']
mini_df.loc[mini_df['fake_freq'] < mini_df['real_freq'], 'fakes'] = ''

mini_df.loc[mini_df['fake_freq'] < mini_df['real_freq'], 'reals'] = mini_df['name']
mini_df.loc[mini_df['fake_freq'] > mini_df['real_freq'], 'reals'] = ''
```

```python
def dict_color_func(word, *args, **kwargs):
    if word in mini_df['fakes'].tolist() :
        color = '#c64039ff'  # red 
    elif word in mini_df['reals'].tolist() :
        color = '#6aa84f' # green 
#     else:
#         color = '#0000' # black other words
    return color
```

```python
fake_cloud = WordCloud(background_color="white",  font_path = '/System/Library/Fonts/HelveticaNeue.ttc',
                       stopwords=my_stopwords, width=800, height=400, color_func = dict_color_func
                         ).generate(' '.join(mini_df['name']))

plt.figure(figsize=(20,10))
# Create a figure of the generated cloud
plt.imshow(fake_cloud, interpolation='bilinear')  
plt.axis('off')
# Display the figure
plt.show()
```

### Simple dictionary BOW / word count filtered labels REAL and FAKE
Filtering the title by label, creating word count dictionaries, and generting word clouds from frequencies. The first cloud we will look at is the most frequent words in the REAL news titles, the second, fake news titles. 

```python
cleaned_fake_split = [x for x in fake_split if x not in my_stopwords]
fake_count = [cleaned_fake_split.count(w) for w in fake_split]
fake_dict = dict(zip(fake_split, fake_count))

cleaned_real_split = [x for x in real_split if x not in my_stopwords]
real_count = [cleaned_fake_split.count(w) for w in real_split]
real_dict = dict(zip(real_split, real_count))
```

```python
# Creating a mask from jpg
REAL_mask = np.array(Image.open("REAL.jpg"))
```

```python
real_cloud = WordCloud(mask =REAL_mask, background_color="white",  font_path = '/System/Library/Fonts/HelveticaNeue.ttc',
                       stopwords=my_stopwords, width=800, height=400,
                         ).generate_from_frequencies(real_dict)

plt.figure(figsize=(20,10))
# Create a figure of the generated cloud
plt.imshow(real_cloud, interpolation='bilinear')  
plt.axis('off')
# Display the figure
plt.show()
```
