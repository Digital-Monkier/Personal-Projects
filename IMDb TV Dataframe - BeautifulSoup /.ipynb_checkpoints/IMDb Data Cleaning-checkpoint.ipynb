# Cardinality reduction
Grouping distributor by platform and LabelEncoding with np.select

```python
df['Distributor'].unique()
```

```python
# CBS will throw up an issue here as it will match in both network and ott lists - unsure if OTT with will overwrite
network_list = ['NBC', 'National Broadcast', 'FOX', 'Fox','ABC', 'American Broadcasting', 
'The CW Network', 'CBS', 'Columbia TriStar Domestic Television', 'Columbia Pictures Television Distribution']


ott_list = ['Netflix', 'Amazon', 'Amazon Studios', 'Amazon Instant Video', 'Disney+', 'CW Seed', 'CBS All Access',
            'HBO Max', 'Apple TV+', 'AMC+', 'Discovery+', 'BET+', 'Amazon Prime Video','Discovery+', 'IMDb',
            'Amazon Instant Video']

p_cable = ['HBO','Home Box Office', 'Showtime']
```

```python
# Label encoding based on list condition
# 1 = Network, 2 = OTT, #3 = Premium Cable, 4 = Basic Cable
# 5 covers None values which appear to be mostly obscure titles - may drop them

df['Dist_num'] = np.select( (df['Distributor'].str.contains('|'.join(network_list)), 
                             df['Distributor'].str.contains('|'.join(ott_list)),
                             df['Distributor'].str.contains('|'.join(p_cable)),
                             df['Distributor'] == 'None'),
                            (1,2,3,5), 4)
```

```python
df[df['Distributor'] == 'CBS All Access']
```

```python
df.loc[df['Distributor'] == 'CBS All Access', 'Dist_num'] = 2
```

# Label Encoding certificates to numeric
Label encoding the the five certificate categories plue one for np.nan

```python
df[df['Certificate'].isna()]
```

```python
# 5 = TVMA, 4 = TV-14, 3 = TV-Y7, 2 = TV-PG, 1 = TV-G
# Creating a mask condition to cover NaN values with 0

mask = df['Certificate'].isna()

df['Cert_num'] = np.select( (df['Certificate'] == 'TV-MA',
                             df['Certificate'] == 'TV-14',
                             df['Certificate'] == 'TV-Y7',
                             df['Certificate'] == 'TV-PG',
                             df['Certificate'] == 'TV-G',
                             df['Certificate'] == mask),
                             (5,4,3,2,1,0))
                             
                             
```

```python
df[df['Certificate'].isna()]
```

# One-Hot-Encoding Genre
* Splitting and extracting the genres from genres column to One-Hot-Encode for modelling later
* Copied IMDb's list of TV Spefific genres

```python
imdb_tvgenres = \
"""Action
Adventure
Animation
Biography
Comedy
Crime
Documentary
Drama
Family
Fantasy
Game-Show
History
Horror
Music
Musical
Mystery
News
Reality-TV
Romance
Sci-Fi
Sport
Talk-Show
Thriller
War
Western"""

genre_split = imdb_tvgenres.split('\n')
```

```python
for genre in genre_split:
    df[genre] = df['Genres'].str.findall(genre)
```

```python
for col in df.iloc[:,18:]:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('[', '').str.replace(']','').str.replace("'", '').str.strip()
```

```python
for col in df.iloc[:,18:]:
    for genre in genre_split:
        if genre in col:
            df[col] = df[col].replace(genre, 1)
        else:
            df[col] = df[col].replace('', 0)
```

```python
df['Music'] = df['Music'].replace('Music, Music', 1)
```

# Filling NaN Runtime Values
* Looking at the runtime values it looks like there are some shows having there whole season counted as runtime 
* Will consider imputing who season runtimes to a mean based on platform and genre at a later stage
* For now will map platform specific medians to the NaN values

```python
dist_group = df.groupby('Dist_num')
medians = dist_group['Runtime (mins)'].median().round().to_dict()
medians
```

```python
mask = df['Runtime (mins)'].isna()
df.loc[mask, 'Runtime (mins)'] = df.loc[mask, 'Dist_num'].map(medians)
```

# Extracting keywords
* Loading the gensim library to extract keywords for vectorization

```python
from gensim.summarization import keywords

df['Keywords'] = df['Description'].apply(lambda x: keywords(\
    x, words=5, split =False, pos_filter=('NNPS', 'NN'))).str.replace('\n', ' ')
```

```python
def distplots(df, col, bins):
    
    fix, ax = plt.subplots(1,3, figsize=(15,5))
    ax[0].hist(df[col],density= True, bins=bins)
    ax[0].plot
    ax[0].set_title(col+' '+'Histogram')
    ax[0].tick_params(labelrotation=45)
    
    ax[1].boxplot(df[col])
    ax[1].plot
    ax[1].set_title(col+' '+'Boxplot')
    
    probplot(df[col], dist='norm', plot = plt)
    ax[2].plot
    ax[2].set_title(col+' '+'Probplot')
    
    
    
```
