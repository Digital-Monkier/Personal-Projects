# Classifier Modelling
* Stacking classiiers
* Dictionary of model, instantiation as key,vale pairs
* Custom scoring to produce classification report for each fold of kfold evaulation function

```python
# Layering some models into the stacking classifier
def stacking():
    # Append tuples of name/model to layer1
    layer1 = list()
    
    # layer1.append(('GNB', GaussianNB())) - Removed from stacking as poorest performer
    layer1.append(('GNB', GaussianNB()))
    layer1.append(('KNC', KNeighborsClassifier()))
    layer1.append(('XGB', XGBClassifier()))
    layer1.append(('DTC', DecisionTreeClassifier()))
    layer1.append(('RFC', RandomForestClassifier()))
    
    layer2 = XGBClassifier()
    
    stacked_model = StackingClassifier(estimators = layer1, final_estimator = layer2, cv=3)
    return stacked_model
```

```python
# Creating a dictionary of model name and instantiations and key, value pairs
def get_models():
    
    models = {}
    
    models['GBC'] = GaussianNB()
    models['KNC'] = KNeighborsClassifier()
    models['XGB'] = XGBClassifier()
    models['DTC'] = DecisionTreeClassifier()
    models['RFC'] = RandomForestClassifier()
    models['STK'] = stacking()
    
    return models
```

```python
# Custom scorer with lists that will extend with each y value provided 
# by cross_val_score which can be spliced later for classification reports of each slice
def kfold_class_report(y, y_pred):
    originalclass.extend(y) # true values 
    predictedclass.extend(y_pred) # predicted values to be initialized prior to model for loop and cleared after 
    return accuracy_score(y, y_pred)
# Generic accuracy score as scoring func must return a number not report when looped 
```

```python
# Evaluation function combines stratified KFold and custom scorer within cross_val_score
def evaluate_models(model, X, y):
    
    cv = StratifiedKFold(n_splits = 3, shuffle =True, random_state = 32)
    score = cross_val_score(model, X, y, scoring= make_scorer(kfold_class_report), n_jobs=1, cv = cv, error_score = 'raise')
    
    return score
```

```python
# Model object
models = get_models()
```

```python
# Class lists to capture the y_true y_pred values
originalclass = []
predictedclass = []

results_train = []
results_test = []
names = []

# For loop to iterate over the models, capture amd visualize the scores
for name, model in models.items():
    #Loading the baseline dataset
    scores_train = evaluate_models(model, X_train, y_train)
    scores_test = evaluate_models(model, X_test, y_test)
    
    results_train.append(scores_train)
    results_test.append(scores_test)
    
    names.append(name)
    print('>%s Train Set Accuracy: %.3f (%.3f)' % (name, np.mean(scores_train), np.std(scores_train))+'\n'\
          '>%s Test Set Accuracy:  %.3f (%.3f)' % (name, np.mean(scores_test), np.std(scores_test))+'\n')

plt.figure(figsize=(10,5))
sns.boxplot(x=names, y=results_train, showmeans=True)
sns.boxplot(x=names, y=results_test, showmeans=True)
```
