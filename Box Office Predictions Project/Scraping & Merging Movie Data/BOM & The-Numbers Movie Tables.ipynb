# Scraping Financials From BOM - 1980-2020

```python
import pandas as pd

# Create A List Of Years as Strings
years = str(list(range(1979, 2022)))
years = years.split(',')
years = [x.strip() for x in years]
years = years[1:42]
```

```python
# Scraping and Appending The Tables On A Loop Using The Variable year as html identifier
import requests

appended_data = pd.DataFrame([])
for year in years:
    url = 'https://www.boxofficemojo.com/year/'+year+'/?grossesOption=calendarGrosses'
    html = requests.get(url).content
    # Creating The Dataframe and Adding The Year As Identifier Column Within Main Dataframe
    df_list = pd.read_html(html)
    dist_df = df_list[-1]
    dist_df['Release Year'] = year
    
    appended_data=appended_data.append(dist_df,ignore_index=True)
    appended_data.to_csv('1980_2020.csv', index=False)
```

```python
# Loading The Dataframe From Newest To Oldest
df = pd.read_csv('1980_2020.csv')
df = df.sort_index(ascending=False)
df = df.drop('Estimated', axis=1)
df.head(2)
```

# Scraping The Distributer Market Share From the-numbers.com

```python
# Using PReloaded Modules Requests / Pandas
url = 'https://www.the-numbers.com/market/distributors'
html = requests.get(url).content
df_list = pd.read_html(html)
dist_df = df_list[-1]
dist_df.to_csv('distributors.csv', index=False)
```

```python
# View Dataframe (NEED TO COMPARE TO  ADD ALT SPELLINGS E.G. Twentith Century & 20th Century)
dist_df = pd.read_csv('distributors.csv')
dist_df.head()
```

```python
# Create A Dictionary From Two Columns of Distributors/Market Share Table To Map To df
distributors = dist_df['Distributor'].tolist()
market_share = dist_df['Share'].tolist()
dist_share_dict = dict(zip(distributors, market_share))
```

```python
# Mapping The Market Share To The BOM 
df['Market Share'] = df['Distributor']
df['Market Share'] = df['Distributor'].map(dist_share_dict)
df.head(10)

# RENOTE: Add The Market Share For Alternative Spellings (Twentith Centrury / 20th Century ) at a later stage
```
